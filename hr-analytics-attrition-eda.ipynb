```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
print(df.shape)
df.head()
```

```python
# Shape
print("Shape:", df.shape)
print("\n")

# Sütun adları
print("Columns:", df.columns.tolist())
print("\n")

# Ümumi məlumat
print(df.info())
print("\n")

# Statistika
df.describe()
```

```python

print(df['Attrition'].value_counts())
print("\n")
print(df['Attrition'].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(6, 4))
df['Attrition'].value_counts().plot(kind='bar',
                                     color=['steelblue', 'salmon'])
plt.title('Employee Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='Age', hue='Attrition',
             bins=30, palette=['steelblue', 'salmon'])
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
dept_attr = df.groupby('Department')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
dept_attr.columns = ['Department', 'Attrition Rate %']

sns.barplot(data=dept_attr, x='Department',
            y='Attrition Rate %', palette='Set2')
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate (%)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome',
            palette=['steelblue', 'salmon'])
plt.title('Monthly Income by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sat_attr = df.groupby('JobSatisfaction')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
sat_attr.columns = ['Job Satisfaction', 'Attrition Rate %']

sns.barplot(data=sat_attr, x='Job Satisfaction',
            y='Attrition Rate %', palette='coolwarm')
plt.title('Attrition Rate by Job Satisfaction (1=Low, 4=High)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
ot_attr = df.groupby('OverTime')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).reset_index()
ot_attr.columns = ['OverTime', 'Attrition Rate %']

sns.barplot(data=ot_attr, x='OverTime',
            y='Attrition Rate %', palette=['steelblue', 'salmon'])
plt.title('Attrition Rate by OverTime')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Yalnız rəqəmsal sütunlar
numeric_df = df.select_dtypes(include=[np.number])

sns.heatmap(numeric_df.corr(),
            annot=False,
            cmap='coolwarm',
            center=0)
plt.title('Correlation Heatmap — HR Analytics')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
sns.boxplot(data=df, x='Attrition',
            y='YearsAtCompany',
            palette=['steelblue', 'salmon'])
plt.title('Years at Company by Attrition')
plt.tight_layout()
plt.show()
```

```python
print("=" * 50)
print("KEY FINDINGS — HR Analytics EDA")
print("=" * 50)

total = len(df)
left = (df['Attrition'] == 'Yes').sum()
rate = left/total*100

print(f"\n1. Overall Attrition Rate: {rate:.1f}%")
print(f"   ({left} out of {total} employees left)\n")

print("2. Avg Age — Left vs Stayed:")
print(df.groupby('Attrition')['Age'].mean().round(1))

print("\n3. Avg Monthly Income — Left vs Stayed:")
print(df.groupby('Attrition')['MonthlyIncome'].mean().round(0))

print("\n4. Avg Years at Company — Left vs Stayed:")
print(df.groupby('Attrition')['YearsAtCompany'].mean().round(1))

print("\n5. Overtime Impact:")
print(df.groupby('OverTime')['Attrition'].apply(
    lambda x: f"{(x=='Yes').sum()/len(x)*100:.1f}%"))
```
